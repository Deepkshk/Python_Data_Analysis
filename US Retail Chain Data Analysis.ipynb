## Let's start Anayse the Data

#### first, we want to know the Total sale

```python
retail['TOT_SALES'].sum().round(2)
```

#### now we want to know the total sale by type of customers

```python
retail.pivot_table(index=['PREMIUM_CUSTOMER'],values=['TOT_SALES'],aggfunc=['sum','mean'])
```

#### here we found Mainstream has the maximum sale.

#### Total sale by Lifestage or age group of customers

```python
retail.pivot_table(index=['LIFESTAGE'],values=['TOT_SALES'],aggfunc=['sum','mean'])
```

#### and here we found Older singles/couples has the maximum sales andnew families has minimum sales

#### Let's check total sale by agegroup and type of customers together

```python
retailgrp=retail.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['TOT_SALES'].sum().unstack()
```

```python
retailgrp
```

#### create a plot to better understand the result

```python
plt.figure(figsize=(20,5))
retailgrp.plot(kind='bar',stacked=True,color=['lightcoral','cyan','limegreen'])
plt.title('Total sale by Customer Lifestage and Premium type')
```

#### we found by plot, New families has min sale in each customer category
#### Let's check Total sale by Brands name

```python
retail.pivot_table(index=['BRAND'],values=['TOT_SALES'],aggfunc=['sum'])
```

#### and here we found Kettle has maximum sale and Burger has minimum sale
#### create a plot for better understanding about Brands sale

```python
sb.barplot(data=retail,x='BRAND',y='TOT_SALES',estimator=sum)
plt.xticks(rotation=90)
```

#### and here we can see the difference between Kettle and Burger Brands
#### Now Let's check Total sale by Brands & Products Name

```python
bppvt=retail.pivot_table(index=['BRAND','PROD_NAME'],values=['TOT_SALES'],aggfunc=['sum'])
bppvt
```

#### and by this result we found Product- Dorito corn chp supreme 380g has maximum sale and woolworths mild salsa 300g has minimum sale.
### for analysis with dates we have to extract day, month and year from DATE column..

```python
retail.head()
```

```python
retail=retail.astype({'DATE':'datetime64[ns]'})
retail.dtypes
```

#### Let's extract the day, month and year

```python
retail['year']=retail['DATE'].dt.year
retail['month']=retail['DATE'].dt.month
```

```python
retail.head()
```

#### Let's check Total sale by year

```python
retail.pivot_table(index=['year'],values=['TOT_SALES'],aggfunc=['sum'])
```

#### and hear we found 2018 (july to december) has more sale than 2019 (january to june)
#### let's deep dive into year with month

```python
retail.pivot_table(index=['year','month'],values=['TOT_SALES'],aggfunc=['sum'])
```

#### and from this result we found December has the maximum sale and february has minimum sale
#### let's create plot to understand more day wise sale 

```python
plt.figure(figsize=(20,8))
sb.lineplot(data=retail,x='DATE',y='TOT_SALES',estimator='sum')
```
